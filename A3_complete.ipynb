```python
# CAREFULLY Select a webpage and retrieve all subdomains
url = "https://www.nytimes.com/"
subdomains = get_subdomains(url)
print(subdomains)
```

```python
# Set parameters for wordcloud
xdim = 1000
ydim = 1000
apply_mask = False
save_fig = True
rstate = 80
contourcolor = 'blue'
contourwidth = 5
maskname = 'mask_w.png'
basename = 'wc_'
imgtype = '.jpg'
maskname = datapath + maskname
location = "America/New_York"
```

```python
#Main loop
#Open each subdomain
#Get the readable text
#Tokenize the text
#Create a unique name
#Create a custom wordcloud with the readable text
#Save the wordcloud image to disk
#------------------------------------------------
limit = 2                    #2 for testing
for i in range (0, limit):
  html = urllib.request.urlopen(subdomains[i]).read()
  webtext = text_from_html(html)
  try:
    token_text = word_tokenize(webtext)
    timestamp = create_timestamp(location)
    savename = datapath + basename + timestamp + '_' + str(i) + imgtype
    print(savename)
    create_wordcloud(token_text, xdim, ydim, rstate, contourcolor, contourwidth, apply_mask, maskname, save_fig, savename)
  except:
    pass
```
